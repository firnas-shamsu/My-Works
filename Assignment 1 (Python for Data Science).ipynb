### Iris-Setosa

```python
Iris_sts = data.loc[:49, ['SL','SW','PL','PW','Classification']]
```

```python
plt.figure(figsize=(6,8))

plt.subplot(2,1,1)
plt.scatter(Iris_sts['SL'], Iris_sts['PL'], s=10)
plt.title('Length of Sepel vs Length of Petel')
plt.xlabel('Sepel Length')
plt.ylabel('Petel Length')

plt.subplot(2,1,2)
plt.scatter(Iris_sts['SW'], Iris_sts['PW'], s=10)
plt.title('Width of Sepel vs Width of Petel')
plt.xlabel('Sepel Width')
plt.ylabel('Petel Width')

plt.tight_layout()
```
